```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

# Initialize Chrome WebDriver without specifying the path
driver = webdriver.Chrome()

# Function to extract information from a Yellow Pages page
def extract_information_from_page(url):
    driver.get(url)
    time.sleep(3)  # Wait for the page to load
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Find all div elements with class "info"
    info_divs = soup.find_all('div', class_='info')
    
    for div in info_divs:
        # Extract name
        name_element = div.find('a', class_='business-name')  # Changed class to 'business-name'
        name = name_element.text.strip() if name_element else 'N/A'

        # Extract category
        category_element = div.find('div', class_='categories')
        category = category_element.text.strip() if category_element else 'N/A'

        # Extract phone number
        phone_element = div.find('div', class_='phones phone primary')
        phone = phone_element.text.strip() if phone_element else 'N/A'

        # Extract address
        address_element = div.find('div', class_='street-address')
        address = address_element.text.strip() if address_element else 'N/A'

        print(f"Name: {name}")
        print(f"Phone: {phone}")
        print(f"Address: {address}")
        print('-' * 30)

# Define the base URL and parameters
base_url = 'https://www.yellowpages.com/search'
search_terms = 'social worker'
geo_location_terms = 'Indianapolis, IN'
num_pages = 5  # Number of pages to scrape

# Loop through pages and extract information
for page_num in range(1, num_pages + 1):
    page_url = f'{base_url}?search_terms={search_terms}&geo_location_terms={geo_location_terms}&page={page_num}'
    print(f"Scraping page {page_num}: {page_url}")
    
    extract_information_from_page(page_url)

# Close the WebDriver
driver.quit()
```
